```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage
```

```python
img = Image.open("../outputs/output_77_2.png")

img_array = np.array(img)

img_array.shape
```

```python
plt.figure(figsize=(12,8))

plt.imshow(img_array)

plt.axis("off")

plt.title("Original Detector Visualization")

plt.show()
```

```python
gray = np.mean(img_array[:,:,:3], axis=2)

plt.figure(figsize=(12,8))

plt.imshow(gray, cmap="inferno")

plt.colorbar(label="Signal Intensity")

plt.title("DPI Observability Heatmap v0.1")

plt.axis("off")

plt.show()
```

```python
sx = ndimage.sobel(gray, axis=0)

sy = ndimage.sobel(gray, axis=1)

sobel = np.hypot(sx, sy)

plt.figure(figsize=(12,8))

plt.imshow(sobel, cmap="magma")

plt.colorbar(label="Structure Gradient")

plt.title("Detector Structure Persistence Map")

plt.axis("off")

plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.imshow(gray, cmap="inferno")

plt.colorbar(label="Signal Intensity")

plt.title("DPI Observability Heatmap v0.1")

plt.axis("off")

plt.savefig(
    "../outputs/overlays/dpi_heatmap_v1.png",
    bbox_inches="tight",
    dpi=300
)

plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.imshow(sobel, cmap="magma")

plt.colorbar(label="Structure Gradient")

plt.title("Detector Structure Persistence Map")

plt.axis("off")

plt.savefig(
    "../outputs/overlays/dpi_structure_map_v1.png",
    bbox_inches="tight",
    dpi=300
)

plt.show()
```

```python
signal_mean = np.mean(gray)

signal_std = np.std(gray)

gradient_mean = np.mean(sobel)

gradient_std = np.std(sobel)

print("=== DPI Observability Metrics ===")
print(f"Signal Mean Intensity: {signal_mean:.2f}")
print(f"Signal Intensity Variability: {signal_std:.2f}")
print(f"Structure Gradient Mean: {gradient_mean:.2f}")
print(f"Structure Gradient Variability: {gradient_std:.2f}")
```
